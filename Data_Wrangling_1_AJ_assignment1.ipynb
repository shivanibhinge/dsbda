# Using Label Encoding

```python
import pandas as pd
```

```python
data_nba = pd.read_csv('/content/drive/MyDrive/nba.csv - nba.csv.csv')
```

```python
from sklearn import preprocessing
```

```python
data_nba['Position'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
data_nba['Position'] = label_encoder.fit_transform(data_nba['Position'])
```

```python
data_nba.head()
```

```python
data_nba['Position'].unique()
```

# Quantitative to Categorical using pandas cut

```python
category = pd.cut(data_nba.Age, bins=[19,25,30,35,45], labels=['A', 'B', 'C', 'D'])
```

```python
data_nba.insert(5, 'Age_Group', category)
```
